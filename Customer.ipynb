**12) Analysing which products drive the most revenue and determine seasonal trend.**

```python
df['revenue']=df['Units Sold']*df['Unit Price']
df['order Date']=pd.to_datetime(df['order Date'] , format='mixed')
df['year_month']=df['order Date'].dt.to_period('M')

monthly_revenue=df.groupby(['year_month','Product Name'])['revenue'].sum().reset_index()
top_10_monthly_products=(monthly_revenue.groupby('Product Name')['revenue'].sum().nlargest(10).index)
monthly_revenue_per_product=monthly_revenue[monthly_revenue['Product Name'].isin(top_10_monthly_products)]
monthly_revenue_pivot=monthly_revenue_per_product.pivot(index='year_month',columns='Product Name',values='revenue').fillna(0)

monthly_revenue_pivot.plot(kind='line', figsize=(9,4),marker='o')
plt.title('Monthly Revenue Per Product')
plt.xlabel('Year-Month')
plt.ylabel('Revenue')
plt.legend(title='Product Name',bbox_to_anchor=(1.05,1),loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

**13) Are there significant difference in average monthly revenue between top selling products.**

```python
import numpy as np
from scipy import stats
monthly_revenue=df.groupby(['year_month','Product Name'])['revenue'].sum().unstack()
product_revenue_data=[monthly_revenue[product].dropna().values for product in monthly_revenue.columns]

f_stat,p_value=stats.f_oneway(*product_revenue_data)

print("ANOVA Test Results")
print("F-statistic:",f_stat)
print("P-value:",p_value)

alpha=0.05
if p_value<alpha:
  print("significant Difference in average monthly revenue between products.")
else:
  print("No significant Difference in average monthly revenue between products.")
```
